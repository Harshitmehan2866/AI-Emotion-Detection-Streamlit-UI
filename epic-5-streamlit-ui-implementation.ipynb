```python
#PAGE CONFIGURATION
import streamlit as st
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

from datetime import datetime
```

```python
#SESSION STATE
if "history" not in st.session_state:
    st.session_state.history=[]

if "results" not in st.session_state:
    st.session_state.results=[]

print("Session Initialized")
```

```python
#SIDEBAR DASHBOARD
st.sidebar.title("Dashboard")

st.sidebar.write(
    f"Total Sessions : {len(st.session_state.history)}"
)

if st.sidebar.button("Clear History"):
    st.session_state.history=[]
    st.session_state.results=[]

st.sidebar.success("Ready")
```

```python
#MAIN LAYOUT
left,right=st.columns([2,1])

with left:
    st.header("Student Learning Assistant")

with right:
    st.info("Emotion Detection Dashboard")
```

```python
#FIELD SELECTION
fields=[
"Computer Science",
"Mathematics",
"Physics",
"Chemistry",
"Biology",
"Engineering",
"Business",
"History",
"Literature",
"Psychology",
"Other"
]

field=st.selectbox(
"Select your field",
fields
)

problem=st.text_area(
"Describe your learning problem",
height=150
)

print(field)
```

```python
#BUTTONS
c1,c2=st.columns(2)

run=c1.button("Analyze")

clear=c2.button("Clear")

if clear:
    problem=""
```

```python
#SPINNER
if run:

    with st.spinner("Analyzing..."):

        import time
        time.sleep(2)

    st.success("Completed")
```

```python
#DUMMY RESULTS
emotion="Confused"
confidence=0.87

bilstm=0.82
bert=0.91

st.session_state.results.append({
    "Emotion":emotion,
    "Confidence":confidence,
    "BiLSTM":bilstm,
    "BERT":bert,
    "Field":field
})
```

```python
#MODEL COMPARISON
if len(st.session_state.results)>0:

    data=st.session_state.results[-1]

    c1,c2=st.columns(2)

    with c1:
        st.metric("BiLSTM",f"{data['BiLSTM']:.2f}")

    with c2:
        st.metric("BERT",f"{data['BERT']:.2f}")
```

```python
#CONFIDENCE BARS
if len(st.session_state.results)>0:

    data=st.session_state.results[-1]

    st.subheader("Confidence")

    st.progress(float(data["BiLSTM"]))

    st.progress(float(data["BERT"]))
```

```python
#MIXED EMOTION DISPLAY
st.subheader("Mixed Emotions")

mixed={
"Confused":0.55,
"Curious":0.28,
"Frustrated":0.17
}

for k,v in mixed.items():
    st.write(k)
    st.progress(v)
```

```python
#ANALYTICS TABS
tab1,tab2,tab3=st.tabs(
[
"Emotions",
"Fields",
"Summary"
]
)

with tab1:
    st.write("Emotion Analytics")

with tab2:
    st.write("Field Analytics")

with tab3:
    st.write("Summary")
```

```python
#PLOTLY PIE CHART
df=pd.DataFrame({
"Emotion":[
"Confused",
"Curious",
"Happy",
"Confused",
"Happy"
]
})

fig=px.pie(
df,
names="Emotion",
title="Emotion Distribution"
)

st.plotly_chart(fig,use_container_width=True)
```

```python
#PLOTLY BAR CHART
df=pd.DataFrame({

"Field":[
"CS",
"Math",
"Physics",
"CS"
],

"Confidence":[
85,
74,
91,
88
]

})

fig=px.bar(
df,
x="Field",
y="Confidence",
title="Confidence by Field"
)

st.plotly_chart(fig,use_container_width=True)
```

```python
#PLOTLY LINE CHART
df=pd.DataFrame({

"Session":[1,2,3,4,5],

"Confidence":[
70,
75,
81,
86,
92
]

})

fig=px.line(
df,
x="Session",
y="Confidence",
markers=True,
title="Confidence Timeline"
)

st.plotly_chart(fig,use_container_width=True)
```

```python
#CACHED FUNCTION
@st.cache_data
def load_data():

    return pd.DataFrame({
        "A":[1,2,3],
        "B":[4,5,6]
    })

df=load_data()

st.dataframe(df)
```

```python
#ERROR HANDLING
try:

    if problem.strip()=="":
        raise Exception("Problem cannot be empty")

    st.success("Input Accepted")

except Exception as e:

    st.error(e)
```
